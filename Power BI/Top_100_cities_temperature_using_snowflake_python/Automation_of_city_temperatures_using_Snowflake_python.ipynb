```python
import pandas as pd
import requests
import re

dd2=pd.DataFrame()
def extract_temperature(text):
    # Remove escape sequence characters from the text
    cleaned_text = re.sub(r'\x1b\[[0-9;]+m', '', text)

    # Find all temperature values in the cleaned text
    temperatures = re.findall(r'\+(\d+)\(\d+\) °C', cleaned_text)

    # Convert temperature values to integers
    temperatures = [int(temp) for temp in temperatures]

    # Find the maximum and minimum temperatures
    max_temp = max(temperatures)
    min_temp = min(temperatures)

    return min_temp, max_temp

today = pd.Timestamp.now().date()
ind = 1

for city in cities:
    # Construct the URL for the city
    url = 'https://wttr.in/{}'.format(city)

    # Send a GET request to fetch the weather data
    res = requests.get(url)

    # Extract temperature from the response text
    temperature = extract_temperature(res.text)

    if temperature:
        min_temp, max_temp = temperature
        print('City: {}'.format(city))
        print('Temperature: Minimum: {}°C, Maximum: {}°C'.format(min_temp, max_temp))
        print('---')
        new_row = {
            'time': today,
            'city': city,
            'tmin': min_temp,
            'tmax': max_temp,
            'date_updated': today,
            'tavg': (min_temp + max_temp) / 2 ,
            'date_updated': today,# Compute the average temperature for each row
        }
        dd2 = dd2.append(new_row, ignore_index=True)
        ind += 1
    else:
        print('Failed to extract temperature for {}'.format(city))
```
